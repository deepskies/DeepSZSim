```python
import simulation as sim
import simulation_tsz2 as sim2
import numpy as np
import os
import shutil
import matplotlib.pyplot as plt

#Load Cosmology
omega_m0, omega_b0, cosmo_h, sigma8, ns = sim2.cosmo_para()

#Prepare directory to store simulated images
cwd = os.getcwd()
directory = 'test'
path = os.path.join(cwd, directory)
shutil.rmtree(path)
os.mkdir(path)

#Select options for images
# nums = np.arange(9) + 1
nums = []

#Input Mass and Redshift
Mass = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
z = np.linspace(0.1,1.5,15)
result = np.arange(10*10*15).reshape(10,10,15)

#Select Telescope and Observation Frequency
tele = 'SPT'
f = 150

#Retrieve Observation-related Parameters
beam_size, noise_level = sim2.tele_para(tele, f)

#Converting to correct scale
f *= 1e9


for k in range(np.size(Mass)):
    for j in range(10):
        for i in range(np.size(z)):
            m = Mass[k] * 1e13
            radius = np.linspace(0.1, 9, 10)     #arcmin
            new_r = sim2.arcmin_to_Mpc(radius, z[i])
            battaglia_pro = np.array([sim2.battaglia_profile(r, m, z[i]) for r in new_r])
            y_pro = sim.epp_to_y(battaglia_pro)
    
            abs_path = os.path.join(path, str(i))
            noise = sim2.generate_img(radius, y_pro, f, noise_level, beam_size, z[i], nums, p = abs_path)
#         print("The noise difference is " + str(noise))
            result[k][j][i] = noise

```

```python
m5 = result[0]
a = m5.mean(axis = 0)
plt.plot(z,a)
```

```python
m5 = result[0]
m10 = result[1]
m15 = result[2]
m20 = result[3]
m25 = result[4]
m30 = result[5]
m35 = result[6]
m40 = result[7]
m45 = result[8]
m50 = result[9]
```

```python
a = m5.mean(axis = 0)
b = m10.mean(axis = 0)
c = m15.mean(axis = 0)
d = m20.mean(axis = 0)
e = m25.mean(axis = 0)
f = m30.mean(axis = 0)
g = m35.mean(axis = 0)
h = m40.mean(axis = 0)
i = m45.mean(axis = 0)
j = m50.mean(axis = 0)
avg = (a+b+c+d+e+f+g+h+i+j)/10
overall_avg = avg.mean()
print("Overall average is: ", overall_avg)
plt.xlabel("Redshift")
plt.ylabel("Noise difference")
plt.scatter(z,a,label = "m = 5 * 1e13")
plt.scatter(z,b,label = "m = 10 * 1e13")
plt.scatter(z,c,label = "m = 15 * 1e13")
plt.scatter(z,d,label = "m = 20 * 1e13")
plt.scatter(z,e,label = "m = 25 * 1e13")
plt.scatter(z,f,label = "m = 30 * 1e13")
plt.scatter(z,g,label = "m = 35 * 1e13")
plt.scatter(z,h,label = "m = 40 * 1e13")
plt.scatter(z,i,label = "m = 45 * 1e13")
plt.scatter(z,j,label = "m = 50 * 1e13")
plt.plot(z,z-z,label = "expected difference")
plt.plot(z,avg,label = "average difference")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
